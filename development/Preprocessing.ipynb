```python
import pandas as pd
import numpy as np

from langdetect import detect
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation

nltk.download('stopwords')
nltk.download('punkt_tab')

sw_indo = set(stopwords.words('indonesian'))
sw_eng = set(stopwords.words('english'))
```

# Merge Dataset

```python
df1 = pd.read_csv("dataset.csv")
df2 = pd.read_csv("dataset2.csv")
df3 = pd.read_csv("dataset3.csv")
df4 = pd.read_csv("dataset4.csv")

df = pd.concat([df1, df2, df3, df4], ignore_index=True)
print(df.shape)
```

# Preprocessing
0. Remove Duplitacate and Null
1. Remove Punction
2. Lowercasing
3. Classify Language
4. Split by Language
5. Stopwords
6. Merge dataset

## 0. Remove Duplicate and Null

```python
print(df.duplicated(subset=["description"]).sum())
```

```python
df.dropna(subset=["description"], inplace=True)
```

```python
df.isna().sum()
```

## 1 2. Remove Punctuation + Lowercasing

```python
punc = set(punctuation)
def remove_punctuation(text):
    return ''.join(word for word in text if word not in punc)

df["description_cleaned"] = df["description"].apply(lambda x : remove_punctuation(str(x).lower()))
df.head()
```

## 3. Classify by Language

```python
df["lang"] = df["description_cleaned"].apply(lambda x : detect(x))
df.head()
```

## 4. Split dataset by Lang 

```python
df_id = df[df["lang"] == "id"]
df_en = df[df["lang"] == "en"]
print(df_id.shape)
print(df_en.shape)
```

## 5. Stopwords

```python
import re

def remove_stopwords(text, sw):
    text = re.sub(r'[^\w\s]', ' ', text)
    tokens = word_tokenize(text)
    return ' '.join([w for w in tokens if w not in sw])

df_id["description_cleaned"] = df_id["description_cleaned"].apply(lambda x : remove_stopwords(x, sw_indo))
df_en["description_cleaned"] = df_en["description_cleaned"].apply(lambda x : remove_stopwords(x, sw_eng))

df = pd.concat([df_id, df_en], ignore_index=True)
df.head()
```

```python
# Save to csv
df.to_csv("dataset_cleaned.csv", index=False)
```
