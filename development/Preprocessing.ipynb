```python
import pandas as pd
import numpy as np

from langdetect import detect
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

## 1 2. Remove Punctuation + Lowercasing

```python
from string import punctuation

punc = set(punctuation)
def remove_punctuation(text):
    return ''.join(word for word in text if word not in punc)

df["description_cleaned"] = df["description"].apply(lambda x : remove_punctuation(str(x).lower()))
df.head()
```

## 3. Classify by Language

```python
df["lang"] = df["description_cleaned"].apply(lambda x : detect(x))
df.head()
```
